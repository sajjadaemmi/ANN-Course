```
import tensorflow as tf
from tensorflow.keras.layers import Dense
from matplotlib import pyplot as plt

dataset = tf.keras.datasets.boston_housing

(x_train, y_train) , (x_test, y_test) = dataset.load_data()
```

```
model = tf.keras.models.Sequential([
        Dense(13, input_dim=13 ,activation="sigmoid"),
        Dense(8, activation="sigmoid"),
        Dense(4, activation="sigmoid"),
        Dense(1, activation="linear"),
])
```

```
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.1, nesterov=False), loss='mse')
```

```
out = model.fit(x_train, y_train, epochs=32)
```

```
plt.plot(out.history['loss'])
plt.show()
```

```
model.evaluate(x_test, y_test)

print(x_test.shape)
```
